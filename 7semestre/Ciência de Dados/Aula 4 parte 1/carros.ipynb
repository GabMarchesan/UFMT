```python
# 游뚱 carros.ipynb - An치lise Explorat칩ria de Dados

# 游닍 Imports iniciais
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from mpl_toolkits.mplot3d import Axes3D

# 游댌 Carregando os dados
df = pd.read_csv("carros_usados.csv", sep=';')

# 游늯 Visualizando as 5 primeiras linhas
df.head()

```

```python
# Estrutura do dataset
print("Shape:", df.shape)
print("\nTipos de dados:")
print(df.dtypes)

# Nomes das colunas
print("\nColunas:")
print(df.columns)

# Verificando valores ausentes
print("\nValores ausentes:")
print(df.isnull().sum())

# Resumo estat칤stico
df.describe()
```

```python
# 3. INFORMA칂칏ES B츼SICAS
# Nome das colunas
print("Colunas do dataset:", df.columns.tolist())

# Verificando se h치 colunas duplicadas
print("Colunas duplicadas:", df.columns[df.columns.duplicated()].tolist())

# Verificando se h치 linhas duplicadas
print(f"Total de linhas duplicadas: {df.duplicated().sum()}")

# Tipos de dados por coluna
print("\nTipos de dados:")
print(df.dtypes)

```

## 2. An치lise Univariada

```python
# Verificando colunas categ칩ricas
cat_cols = df.select_dtypes(include='object').columns.tolist()
print("Vari치veis Categ칩ricas:", cat_cols)

# Frequ칡ncia das categorias
for col in cat_cols:
    print(f"\nColuna: {col}")
    print(df[col].value_counts())
    df[col].value_counts().plot(kind='bar')
    plt.title(col)
    plt.show()
```

```python
# Verificando colunas num칠ricas
num_cols = df.select_dtypes(include=np.number).columns.tolist()
print("Vari치veis Num칠ricas:", num_cols)

# Histograma das vari치veis num칠ricas
df[num_cols].hist(bins=20, figsize=(15, 10))
plt.suptitle("Distribui칞칚o das vari치veis num칠ricas")
plt.show()
```

## An치lise Bivariada

```python
# Selecionar vari치veis num칠ricas
numericas = df.select_dtypes(include=['int64', 'float64']).columns

# Scatter plots
plt.figure(figsize=(12, 5))
sns.scatterplot(x='mileage', y='lot.sale.days', data=df)
plt.title("Quilometragem vs Tempo de Venda")
plt.show()

sns.scatterplot(x='vehicle.age', y='total.cost', data=df)
plt.title("Idade do Ve칤culo vs Custo Total")
plt.show()

# Pairplot com vari치veis num칠ricas
sns.pairplot(df[numericas].dropna())
plt.suptitle("Pairplot das Vari치veis Num칠ricas", y=1.02)
plt.show()
```

## 4. An치lise Multivariada

```python
# Heatmap de correla칞칚o
plt.figure(figsize=(10, 8))
correlacao = df[numericas].corr()
sns.heatmap(correlacao, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Heatmap de Correla칞칚o")
plt.show()

# Gr치fico 3D - Quilometragem x Idade x Custo
fig = px.scatter_3d(df, x='mileage', y='vehicle.age', z='total.cost',
                    color='vehicle.type', title="3D - Mileage vs Age vs Total Cost")
fig.show()
```
